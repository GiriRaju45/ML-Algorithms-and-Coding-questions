### Linear Regression

Linear Regression involves fitting a linear equation or straight line on the given data. It assumes, there is a linear relationship between the dependent(X) variable and the independent(Y) variable in the data. 

It can denoted by the equation:

                        y^ = b0 + b1*x + e  i.e  y^ = m*x + c

        where, b0 and b1 are the coefficients of x that we need to estimate. 


least square is the method that we can use to find the best fit line. least square method involves in reducing the sum of square of the vertical discrepancy between the line and the data points to be small as much as possible.

    s = (y - y^)^2  --> minimum, where y^ = b0 + b1*x 

we need to find b0 and b1 such that the difference between the data point y(i) of (x(i), y(i)) and the line y is minimum.  

   by performing diffrentiation on s we can able to find the value of b0, b1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
# Loading the data

df = pd.read_csv('/Users/guest1/Downloads/archive/Salary_Data.csv')
df.head()
print(df.columns)
print('No.of entries in the dataset: ' ,len(df))

plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Exp.')
plt.ylabel('Salary')
plt.show()
```

```python
x = np.array(df['YearsExperience'])
y = np.array(df['Salary'])

mean_x = np.mean(x)
mean_y = np.mean(y)

print(mean_x, mean_y)
max(x)
```

```python
def covariance(x, mean_x, y, mean_y):
    covar = 0.0
    for i in range(len(x)):
        covar += (x[i] - mean_x) * (y[i] - mean_y)
    return covar
 
# Calculate the variance of a list of numbers
def variance(values, mean):
 return sum([(x-mean)**2 for x in values])


print(covariance(x.tolist(), mean_x, y.tolist(), mean_y))
print(np.shape(np.cov(x,y)))
```

```python
b1 = covariance(x, mean_x, y, mean_y) / variance(x, mean_x)


b0 = mean_y - np.dot(b1,mean_x)
y_hat = []
for x_ in x:
    pred_x = round(b0 + b1*x_, 2)
    y_hat.append(pred_x)
len(y_hat)

print(y_hat)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(df['YearsExperience'], y_hat)
plt.xlabel('Years of Exp.')
plt.ylabel('predicted_salary')
plt.show()
```

### Linear Regeression using Sci-kit learn:


```python
from sklearn.linear_model import LinearRegression


```

```python
x_train = list(map(lambda x: [x], x))
y_train = list(map(lambda x: [x], y))
x_test = list(map(lambda x: [x], np.round(np.random.uniform(0,11, 30),2)))
```

```python
linreg = LinearRegression()

model = linreg.fit(x_train, y_train)
pred = model.predict(x_test)

# Plotting the predictions
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(x_test, pred)
plt.xlabel('Years of Exp.')
plt.ylabel('predicted_salary')
plt.show()
```
